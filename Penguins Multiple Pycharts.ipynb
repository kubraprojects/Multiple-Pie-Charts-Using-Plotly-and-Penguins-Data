```python
#specieler gruplandı

species_data = df.groupby('species')

pie_charts = []

#renk seç ve ters çevir, çalışıyor ancak oynama yapınca kodun içindeki ismi değişmeyi unutma
#niye ters çeviriyoz: normalde büyük dilim en sondaki rengi alıyor diye sevmedim

set1_palette= px.colors.qualitative.Pastel[5:8]

palette = set1_palette[::-1]

for species, data in species_data:
    mean_values = data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']].mean()
    labels = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']
    values = mean_values.values

    pie_chart = go.Figure(data=[go.Pie(labels=labels,values=values, 
                                       title=species, title_font=dict(size=20),
                                       textfont=dict(size=14),marker=dict(colors=palette))]) #textfont:pie içindeki sayıların boyu
    
    pie_chart.update_layout(title=species)
    pie_charts.append(pie_chart)
    
fig = make_subplots(rows=1, cols=len(pie_charts), 
                    subplot_titles=[" "], specs=[[{'type': 'pie'}] * len(pie_charts)]) #iki tane tür başlığı oluşuyordu " " yaptım

for i, chart in enumerate(pie_charts, start=1):
    fig.add_trace(chart.data[0], row=1, col=i)

fig.update_layout(height=700, width=1100, title="Interactive Dynamic Subplot Pie Chart", showlegend=True,legend_font=dict(size=20) )

chart_studio.tools.set_credentials_file(username=username, api_key=api_key)
py.plot(fig, filename = 'Subplot Pie Chart', auto_open=True)
```
